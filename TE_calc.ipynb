```python
import numpy as np
```

```python
def H_conc(K, c):
    d = K**2 + 4*c*K
    return (-K + np.sqrt(d)) / 2
```

```python
# Opgave 2

print('a)')
H = H_conc(5.00*10**(-6), 0.02)
print(f'[H+] = {H} M')
pH = -np.log10(H)
print(f'pH = {pH}')

print('b)')
A = H
pOH = 14-pH
print(f'pOH = {pOH}')
OH = 10**(-pOH)
print(f'[OH-] = {OH} M')

print('c)')
A_after = (5.00*10**(-6)*0.02) / 0.20
print(f'[A-] = {A_after} M')
```

```python
# Opgave 3

print('a)')
H = H_conc(4.5*10**(-4), 0.250)
print(f'[H+] = {H} M')
pH = -np.log10(H)
print(f'pH = {pH}')

print('b)')
Kb = 10**(-14) / (4.5*10**(-4))
print(f'Kb = {Kb}')
c_NaNO2 = 0.005 / 0.075
print(f'c(NaNO2) = {c_NaNO2} M')
OH = H_conc(Kb, c_NaNO2)
print(f'[OH-] = {OH} M')
pH = 14 + np.log10(OH)
print(f'pH = {pH}')
```
